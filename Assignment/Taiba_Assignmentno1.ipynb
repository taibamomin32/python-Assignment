```python
#5.Take a year as input and print whether it is a leap year or not.
```

```python
#6. Build a simple calculator take two numbers and operator(+,-,*,/)as input perform the operation and print the result
#print an error message for invalid operators or division by zero 
```

```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
operator = input("Enter operator (+, -, *, /): ")


if operator == "+":
    result = num1 + num2
elif operator == "-":
    result = num1 - num2
elif operator == "*":
    result = num1 * num2
elif operator == "/":
    if num2 != 0:
        result = num1 / num2
    else:
        result = "Error! Division by zero."
else:
    result = "Invalid operator."

print("Result:", result)
```

```python
#9. keep accepting numbers from the user until they enter 0 then print the total count of numbers entered their sum,and their average.9
```

```python
count = 0
total = 0

while True:
    num = float(input("Enter a number (0 to stop): "))

    if num == 0:
        break

    count += 1
    total += num

if count > 0:
    average = total / count
    print("Count:", count)
    print("Sum:", total)
    print("Average:", average)
else:
    print("No numbers were entered.")
```

```python
#10. take n as input and print a triangle pattern of n rows made of asterisks ,where row i has (2i-1) stars centered appropriately.
```

```python
n = int(input("Enter number of rows: "))

for i in range(1, n + 1):
    spaces = n - i
    stars = 2 * i - 1
    print(" " * spaces + "*" * stars)
    
```
