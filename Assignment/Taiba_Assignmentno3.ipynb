```python
Q1.BankAccount class
```

```python
class BankAccount:
    def __init__(self, holder, acc_no, balance):
        self.account_holder = holder
        self.account_number = acc_no
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient Balance")
        else:
            self.balance -= amount

    def get_balance(self):
        return self.balance

    def __str__(self):
        return f"Holder: {self.account_holder}, Account No: {self.account_number}, Balance: {self.balance}"

a1 = BankAccount("Ali", 101, 5000)
a2 = BankAccount("Sara", 102, 3000)

a1.deposit(1000)
a1.withdraw(2000)

print(a1)
print(a2)
```

```python
Q2.Student class
```

```python
class Student:
    def __init__(self, name, roll_no, marks):
        self.name = name
        self.roll_no = roll_no
        self.marks = marks

    def calculate_percentage(self):
        return sum(self.marks) / len(self.marks)

    def get_grade(self):
        per = self.calculate_percentage()
        if per >= 80:
            return "A"
        elif per >= 60:
            return "B"
        elif per >= 40:
            return "C"
        else:
            return "Fail"

    def display(self):
        print(self.name, self.roll_no,
              self.calculate_percentage(),
              self.get_grade())

s1 = Student("A", 1, [80,85,90,75,88])
s2 = Student("B", 2, [70,65,60,75,68])
s3 = Student("C", 3, [35,40,30,45,38])

s1.display()
s2.display()
s3.display()
```

```python
Q3.Employee (Encasulation)
```

```python
class Employee:
    def __init__(self, name, emp_id, salary):
        self.__name = name
        self.__emp_id = emp_id
        self.__salary = salary

    def get_salary(self):
        return self.__salary

    def set_salary(self, salary):
        if salary < 10000:
            raise ValueError("Salary below 10000")
        self.__salary = salary

    def give_raise(self, percent):
        self.__salary += self.__salary * percent / 100

e = Employee("Ahmed", 1, 15000)
e.give_raise(10)
print(e.get_salary())
```

```python
Q4.SecureVault
```

```python
class SecureVault:
    def __init__(self, pin):
        self.__pin = pin
        self.__data = {}

    def add_entry(self, pin, key, value):
        if pin == self.__pin:
            self.__data[key] = value
        else:
            print("Access Denied")

    def get_entry(self, pin, key):
        if pin == self.__pin:
            return self.__data.get(key)
        print("Access Denied")


v = SecureVault(1234)
v.add_entry(1234, "Name", "Taiba")
print(v.get_entry(1234, "Name"))
```

```python
Q5.shape inheritance
```

```python
import math

class Shape:
    def area(self):
        return 0

    def describe(self):
        print("I am a Shape")

class Circle(Shape):
    def __init__(self, r):
        self.r = r

    def area(self):
        return math.pi * self.r**2

    def describe(self):
        print("I am a Circle")

class Rectangle(Shape):
    def __init__(self, l, w):
        self.l = l
        self.w = w

    def area(self):
        return self.l * self.w

    def describe(self):
        print("I am a Rectangle")

class Triangle(Shape):
    def __init__(self, b, h):
        self.b = b
        self.h = h

    def area(self):
        return 0.5 * self.b * self.h

    def describe(self):
        print("I am a Triangle")
    
```

```python
Q6.vehicle
```

```python
class Vehicle:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def start_engine(self):
        print("Engine started")

class Car(Vehicle):
    def __init__(self, make, model, year, doors):
        super().__init__(make, model, year)
        self.doors = doors

    def start_engine(self):
        print(f"{self.model} Engine started")

class ElectricCar(Vehicle):
    def __init__(self, make, model, year, battery):
        super().__init__(make, model, year)
        self.battery = battery

    def start_engine(self):
        print(f"Silent start - battery at {self.battery} kWh")
        
```

```python
Q7.Polymorphism
```

```python
class Animal:
    def sound(self):
        print("Some sound")

class Dog(Animal):
    def sound(self):
        print("Bark")

class Cat(Animal):
    def sound(self):
        print("Meow")

class Cow(Animal):
    def sound(self):
        print("Moo")

def make_all_sounds(animals):
    for a in animals:
        a.sound()

animals = [Dog(), Cat(), Cow()]
make_all_sounds(animals)
```

```python
Q8.calculator and vector
```

```python
class Calculator:
    def add(self, a, b, c=0):
        return a + b + c

    def multiply(self, a, b, c=1):
        return a * b * c

class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"({self.x}, {self.y})"

v1 = Vector(2, 3)
v2 = Vector(4, 5)

print(v1 + v2)
```

```python
#### class StudentManager:
    def __init__(self):
        self.students = []

    def add_student(self, name, roll, marks):
        self.students.append([name, roll, marks])

    def view_all(self):
        for s in self.students:
            avg = sum(s[2]) / len(s[2])
            print(s[0], s[1], avg)

sm = StudentManager()

while True:
    print("1.Add 2.View 3.Exit")
    ch = input("Choice: ")

    if ch == "1":
        name = input("Name: ")
        roll = input("Roll: ")
        marks = list(map(int, input("Marks: ").split()))
        sm.add_student(name, roll, marks)

    elif ch == "2":
        sm.view_all()

    elif ch == "3":
        break

    else:
        print("Invalid Choice")
        
```
