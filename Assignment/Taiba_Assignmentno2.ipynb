```python
Q1. Write a function analyze_sentence(sentence) that returns dict with word_count, char_count, unique_words, most_frequent_word
```

```python
def analyze_sentence(sentence):
    words = sentence.lower().split()
    word_count = len(words)
    char_count = len(sentence.replace(" ", ""))
    unique_words = len(set(words))

    freq = {}
    for word in words:
        freq[word] = freq.get(word, 0) + 1
    most_frequent_word = max(freq, key=freq.get)

    return {
        "word_count": word_count,
        "char_count": char_count,
        "unique_words": unique_words,
        "most_frequent_word": most_frequent_word
    }

# Test
print(analyze_sentence('The quick brown fox jumps over the lazy dog the fox'))
# Output: {'word_count': 11, 'char_count': 36, 'unique_words': 9, 'most_frequent_word': 'the'}
```

```python
Q2. Given text = 'A man a plan a canal Panama'
```

```python
text = 'A man a plan a canal Panama'

# a) Reverse without slicing
rev = ""
for ch in text:
    rev = ch + rev
print("a) Reverse:", rev)

# b) Palindrome check ignoring case & spaces
clean = text.lower().replace(" ", "")
print("b) Palindrome:", clean == clean[::-1])

# c) Replace every vowel with *
for v in 'aeiouAEIOU':
    text = text.replace(v, '*')
print("c) Vowels replaced:", text)

# d) Count vowels
text2 = 'A man a plan a canal Panama'
vowel_count = {v: text2.lower().count(v) for v in 'aeiou'}
print("d) Vowel count:", vowel_count)
```

```python
Q3. format_report(data) - name left 15, score right 6, grade center 
```

```python
def format_report(data):
    print(f"{'Name':<15} {'Score':>6} {'Grade':^8}")
    print("-" * 30)
    for name, score, grade in data:
        print(f"{name:<15} {score:>6} {grade:^8}")

data = [('Alice', 92, 'A'), ('Bob', 78, 'B+'), ('Charlie', 55, 'C'), ('Diana', 100, 'A+')]
format_report(data)
```

```python
Q4. list_ops(nums) on nums = [3][1][4][1][5][9][2][6][5][3][5] and nested = [[1][2][3],[4][5],[6][7][8][9]]
```

```python
def list_ops(nums, nested):
    # a) Remove duplicates preserving order
    no_dup = list(dict.fromkeys(nums))

    # b) Square elements at even indices
    squared = [nums[i]**2 for i in range(0, len(nums), 2)]

    # c) Flatten nested list
    flat = [x for sublist in nested for x in sublist]

    return no_dup, squared, flat

nums = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5]
nested = [[1, 2, 3], [4, 5], [6, 7, 8, 9]]
print("a) No dup:", list_ops(nums, nested)[0])
print("b) Squared even indices:", list_ops(nums, nested)[1])
print("c) Flattened:", list_ops(nums, nested)[2])
```

```python
Q7. inventory_manager(records)
```

```python
def inventory_manager(records):
    result = {}
    for product, cat, qty, price in records:
        if cat not in result:
            result[cat] = {"total_items": 0, "total_value": 0, "products": []}
        result[cat]["total_items"] += qty
        result[cat]["total_value"] += qty * price
        result[cat]["products"].append(product)
    return result

records = [('Laptop', 'Electronics', 10, 55000), ('Mouse', 'Electronics', 50, 499),
           ('Desk', 'Furniture', 8, 12000), ('Chair', 'Furniture', 15, 4500),
           ('Pen', 'Stationery', 200, 20)]
print(inventory_manager(records))
```

```python
Q8. a) power_all(_args, exp=2) b) profile(*kwargs) c) sort by cgpa desc_
```

```python

def power_all(*args, exp=2):
    return [x**exp for x in args]
print("a) power_all:", power_all(2, 3, 4, exp=3))

# b)
def profile(**kwargs):
    for k, v in kwargs.items():
        print(f"{k}: {v}")
print("b) profile:")
profile(name="Riya", age=20, city="Pune")

# c)
students = [{'name': 'Arjun', 'cgpa': 8.4}, {'name': 'Sanya', 'cgpa': 9.1}, {'name': 'Dev', 'cgpa': 7.8}]
sorted_students = sorted(students, key=lambda x: x['cgpa'], reverse=True)
print("c) Sorted:", sorted_students)
```

```python
Q9. stats_utils.py with mean, variance, normalize
stats_utils.py file me ye code:
```

```python
def mean(data):
    return sum(data) / len(data)

def variance(data):
    m = mean(data)
    return sum((x - m)**2 for x in data) / len(data)

def normalize(data):
    mn, mx = min(data), max(data)
    return [(x - mn) / (mx - mn) for x in data]
```

```python
Q10. a) Closure b) @timer decorator c) @validate_positive decorator
```

```python
import time

# a) Closure
def make_multiplier(n):
    def multiplier(x):
        return x * n
    return multiplier

triple = make_multiplier(3)
quintuple = make_multiplier(5)
print("a) triple(10):", triple(10), "quintuple(10):", quintuple(10))

# b) Timer decorator
def timer(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        print(f"b) Time taken: {time.time() - start:.4f}s")
        return result
    return wrapper

@timer
def slow_sum(n):
    return sum(i*i for i in range(1, n+1))

slow_sum(10000)

# c) Validate decorator
def validate_positive(func):
    def wrapper(x):
        if x <= 0:
            raise ValueError("x must be positive")
        return func(x)
    return wrapper

@validate_positive
def safe_sqrt(x):
    return x ** 0.5

print("c) safe_sqrt(25):", safe_sqrt(25))
```
