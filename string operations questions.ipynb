```python
vowel counter
```

```python
def vowel_counter(s):
    vowels = "aeiou"
    counts = {}

    for ch in s.lower():
        if ch in vowels:
            counts[ch] = counts.get(ch, 0) + 1

    print("Total vowels:", sum(counts.values()))
    print("Vowel counts:", counts)

vowel_counter("Hello World")
```

```python
palindrome cheker
```

```python
def palindrome_checker(s):
    rev = []

    for ch in s:
        rev.insert(0, ch)

    rev_str = "".join(rev)

    if s == rev_str:
        print("Palindrome")
    else:
        print("Not Palindrome")

palindrome_checker("madam")
```

```python
3.character frequency counter
```

```python
def char_frequency(s):
    freq = {}

    for ch in s:
        if ch in freq:
            freq[ch] += 1
        else:
            freq[ch] = 1

    print(freq)

char_frequency("hello")
```

```python
4.word length classification
```

```python
def classify_words(sentence):
    short_words = []
    medium_words = []
    long_words = []

    for word in sentence.split():
        if len(word) < 5:
            short_words.append(word)
        elif len(word) <= 8:
            medium_words.append(word)
        else:
            long_words.append(word)

    print("Short:", short_words)
    print("Medium:", medium_words)
    print("Long:", long_words)

classify_words("Python is a powerful programming language")
```

```python
5.substrings with specific start/end
```

```python
def matching_strings(lst):
    result = []

    for s in lst:
        if s.startswith("py") and s.endswith("thon"):
            result.append(s)

    print(result)

matching_strings(["python", "pythons", "python"])
```

```python
6.unique word extractor
```

```python
import string

def unique_words(paragraph):
    paragraph = paragraph.lower()

    for p in string.punctuation:
        paragraph = paragraph.replace(p, "")

    words = paragraph.split()
    unique = set(words)

    print(sorted(unique))

unique_words("Hello world! Hello Python.")
```

```python
7.even/odd indexd character separator
```

```python
def even_odd_chars(s):
    even = []
    odd = []

    for i, ch in enumerate(s):
        if i % 2 == 0:
            even.append(ch)
        else:
            odd.append(ch)

    print("Even:", even)
    print("Odd:", odd)

even_odd_chars("Python")
```

```python
8.string anagram checker
```

```python
def anagram_check(s1, s2):
    if sorted(list(s1)) == sorted(list(s2)):
        print("Anagram")
    else:
        print("Not Anagram")

anagram_check("listen", "silent")
```

```python
9.missing character finder
```

```python
def missing_letters(s):
    letters = set("abcdefghijklmnopqrstuvwxyz")

    for ch in s.lower():
        if ch in letters:
            letters.remove(ch)

    print("Missing letters:", letters)

missing_letters("hello world")
```

```python
10.custom string reversal(word by word)
```

```python
def reverse_words(sentence):
    words = sentence.split()

    reversed_words = words[::-1]

    result = " ".join(reversed_words)

    print(result)

reverse_words("Hello World")
```
