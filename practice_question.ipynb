```python
# 2: Sum of Even Numbers
```

```python
numbers = [12, 7, 23, 48, 15, 66, 9]

sum_even = 0

for num in numbers:
    if num % 2 == 0:
        sum_even += num

print("Sum of even numbers:", sum_even)
```

```python
#3: Manual List Reversal
```

```python
data = ['apple', 'banana', 'cherry', 'date', 'elderberry']

reversed_list = []

for i in range(len(data)-1, -1, -1):
    reversed_list.append(data[i])

print(reversed_list)
```

```python
# 4: Finding the Maximum Element
```

```python
values = [8, 2, 10, 5, 17, 3, 14]

largest = values[0]

for num in values:
    if num > largest:
        largest = num

print("Largest number:", largest)
```

```python
# 5: Count Occurrences
```

```python
items = ['red', 'blue', 'green', 'red', 'yellow', 'blue', 'red']

count = 0

for color in items:
    if color == 'red':
        count += 1

print("Count of red:", count)
```

```python
# 6: Remove All Occurrences of an Element
```

```python
mixed_numbers = [1, 5, 2, 5, 3, 5, 4, 5]

new_list = []

for num in mixed_numbers:
    if num != 5:
        new_list.append(num)

print(new_list)
```
